```r
# Pour exécuter ce code, assurez-vous que votre noyau est R ou utilisez %%R avec rpy2

# Générer un échantillon de données (simulées pour l'exercice)
taille_echantillon <- c(175, 182, 169, 171, 178, 185, 163, 176, 188, 170)

# Calculer la moyenne de l'échantillon
moyenne_estimee <- mean(taille_echantillon)
print(paste("L'estimation ponctuelle de la taille moyenne est :", moyenne_estimee))

# Calculer la médiane comme autre estimateur de la tendance centrale
mediane_estimee <- median(taille_echantillon)
print(paste("L'estimation ponctuelle de la taille médiane est :", mediane_estimee))
```

```r
# Notes d'un échantillon d'étudiants
notes_examen <- c(85, 92, 78, 88, 95, 81, 75, 89, 93, 80)

# Calculer la variance de l'échantillon (non biaisée)
# La fonction var() en R calcule par défaut l'estimateur non biaisé (division par n-1)
variance_estimee <- var(notes_examen)
print(paste("L'estimation ponctuelle de la variance des notes est :", variance_estimee))
```

```r
# ----------------------------
# PARAMÈTRES
# ----------------------------
Nsim <- 10000
n <- 30

# --- Loi normale ---
mu <- 5
sigma2 <- 4
sigma <- sqrt(sigma2)

# --- Loi exponentielle ---
lambda_true <- 2


# ==============================
# 1) LOI NORMALE
# ==============================
means_normal <- numeric(Nsim)

for (i in 1:Nsim) {
  sample <- rnorm(n, mean = mu, sd = sigma)
  means_normal[i] <- mean(sample)
}

var_theo_normal <- sigma2 / n

cat("=== LOI NORMALE ===\n")
cat("Valeur théorique de µ :", mu, "\n")
cat("Moyenne empirique des estimateurs :", mean(means_normal), "\n")
cat("Variance théorique :", var_theo_normal, "\n")
cat("Variance empirique :", var(means_normal), "\n\n")


# ==============================
# 2) LOI EXPONENTIELLE
# ==============================
lambdas_exp <- numeric(Nsim)

for (i in 1:Nsim) {
  sample <- rexp(n, rate = lambda_true)
  lambdas_exp[i] <- 1 / mean(sample)  # EMV
}

var_theo_exp <- lambda_true^2 / n

cat("=== LOI EXPONENTIELLE ===\n")
cat("Valeur théorique de λ :", lambda_true, "\n")
cat("Moyenne empirique des estimateurs :", mean(lambdas_exp), "\n")
cat("Variance théorique :", var_theo_exp, "\n")
cat("Variance empirique :", var(lambdas_exp), "\n")
```
